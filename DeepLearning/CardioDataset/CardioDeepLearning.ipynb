```python
X = data.loc[:,'age':'active':1]
X
```

```python
transformed_x = minmax_model.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(transformed_x,Y,shuffle=True,random_state=79,test_size=0.3)
```

```python
Y = data.loc[:,'cardio']
Y
```

```python
transformed_y = pd.get_dummies(Y)
transformed_y
```

```python
x_train_ohe,x_test_ohe,y_train_ohe,y_test_ohe = train_test_split(transformed_x,transformed_y,shuffle=True,random_state=79,test_size=0.3)
```

```python
model_ohe = tf.keras.models.Sequential([
    tf.keras.layers.InputLayer(input_shape=11),
    tf.keras.layers.Dense(40,activation='relu'),
    tf.keras.layers.Dense(20,activation='relu'),
    tf.keras.layers.Dense(10,activation='relu'),
    tf.keras.layers.Dense(2,activation="softmax")
])
```

```python
model_ohe.compile(optimizer=tf.keras.optimizers.Adam(0.001),loss=tf.keras.losses.CategoricalCrossentropy(),metrics=tf.keras.metrics.CategoricalAccuracy())
```

```python
model_ohe.fit(x_train_ohe,y_train_ohe,epochs=50,validation_data=(x_test_ohe,y_test_ohe),batch_size=100)
```
